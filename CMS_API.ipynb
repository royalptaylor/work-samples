```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
```

```python
end_point = 'https://data.medicare.gov/resource/kac9-a9fp.json'
```

```python
response = requests.get(end_point).json()
```

```python
response
```

```python
response_df = pd.DataFrame(response)
```

```python
response_df
```

```python
response_df.loc[response_df['state'] == 'CO']
```

```python
drop_index = response_df[response_df['readm_ratio'] == 'Not Available'].index
cleaned_response_df = response_df.drop(drop_index)
```

```python
drop_index2 = cleaned_response_df[cleaned_response_df['number_of_discharges'] == 'Not Available'].index
cleaned_response_df = cleaned_response_df.drop(drop_index2)
```

```python
drop_index3 = cleaned_response_df[cleaned_response_df['number_of_readmissions'] == 'Not Available'].index
cleaned_response_df = cleaned_response_df.drop(drop_index3)
```

```python
cleaned_response_df
```

```python
cleaned_response_df['Colorado'] = np.where(cleaned_response_df['state'] == 'CO', 1, 0)
```

```python
cleaned_response_df
```

```python
cleaned_response_df.info()
```

```python
info_url = 'https://data.medicare.gov/resource/rbry-mqwu.json'
```

```python
info_response = requests.get(info_url).json()
info_response
```

```python
info_df = pd.DataFrame(info_response)
info_df.info()
```

```python
info_df_reduced = info_df[['provider_id', 'hospital_name', 'address', 'city', 'state']].copy()
```

```python
readmin_df = pd.merge(cleaned_response_df, info_df_reduced, on='provider_id')
readmin_df.info()
```

```python
 cleaned_response_df.to_csv('readmissions.csv')
```
