### Operational Efficiency Example

Due to access constraints within our enterprise resource planning system, members of my team were manually emailing in upwards of 300 ad-hoc reports to physicians and researchers on a monthly basis. This manual process took between 20 and 30 person-hours per month. 

I developed this workbook to automatically email these reports using a team member's Microsoft Outlook log-in credentials.

```python
import win32com.client as win32
import pandas as pd
import numpy as np
```

```python
# create email subject and body
subject = 'May 2019 Fund Reports'
body = 'Please find attached your May 2019 fund balance reports.'
```

```python
# import email address and attachment file path data
df_test = pd.read_excel('example.xlsx')
df_test
```

```python
# send emails
for index, row in df_test.iterrows():
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.From = row['From Address']
    mail.To = row['Email Address']
    mail.cc = row['Admin Email Address']
    mail.Subject = subject
    mail.Body = body
    
    # To attach a file to the email:
    attachment = row['Attachment']
    mail.Attachments.Add(attachment)
    mail.Send()
```
